```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# ===============================
# FILE
# ===============================
CSV_FILE = "/home/meenakshi/bragg-peak/build/detector_hits_10e8_2.csv"

# ===============================
# COLUMN INDICES (0-based)
# ===============================
HIT_Z_COL  = 5     # hit Z [mm]
HIT_E_COL  = 6     # hit energy [MeV]
TRUE_Z_COL = 10    # true production Z [mm]
TRUE_E_COL = 11    # true energy [MeV]

USECOLS = [HIT_Z_COL, HIT_E_COL, TRUE_Z_COL, TRUE_E_COL]

# ===============================
# ENERGY WINDOW
# ===============================
E_MIN = 4.0
E_MAX = 5.0

# ===============================
# DEPTH RANGE & BINNING
# ===============================
Z_MIN = -300
Z_MAX = 300
BIN_WIDTH = 10.0   # mm

# ===============================
# STORAGE
# ===============================
z_hit_all  = []
z_hit_45   = []
z_true_all = []
z_true_45  = []

# ===============================
# READ DATA (CHUNKED PANDAS)
# ===============================
for chunk in pd.read_csv(
    CSV_FILE,
    usecols=USECOLS,
    chunksize=1_000_000
):
    zh = chunk.iloc[:, 0].values
    Eh = chunk.iloc[:, 1].values
    zt = chunk.iloc[:, 2].values
    Et = chunk.iloc[:, 3].values

    # HIT Z
    mask_hit_z = (zh >= Z_MIN) & (zh <= Z_MAX)
    zh = zh[mask_hit_z]
    Eh = Eh[mask_hit_z]

    z_hit_all.append(zh)
    z_hit_45.append(zh[(Eh >= E_MIN) & (Eh <= E_MAX)])

    # TRUE Z
    mask_true_z = (zt >= Z_MIN) & (zt <= Z_MAX)
    zt = zt[mask_true_z]
    Et = Et[mask_true_z]

    z_true_all.append(zt)
    z_true_45.append(zt[(Et >= E_MIN) & (Et <= E_MAX)])

# Concatenate chunks
z_hit_all  = np.concatenate(z_hit_all)
z_hit_45   = np.concatenate(z_hit_45)
z_true_all = np.concatenate(z_true_all)
z_true_45  = np.concatenate(z_true_45)

print(f"Hit all gammas      : {len(z_hit_all):,}")
print(f"Hit 4–5 MeV gammas  : {len(z_hit_45):,}")
print(f"True all gammas     : {len(z_true_all):,}")
print(f"True 4–5 MeV gammas : {len(z_true_45):,}")

# ===============================
# ALIGN ALL TO SAME ENTRANCE
# ===============================
z0 = min(z_hit_all.min(), z_true_all.min())

z_hit_all  -= z0
z_hit_45   -= z0
z_true_all -= z0
z_true_45  -= z0

# ===============================
# COMMON HISTOGRAM BINS
# ===============================
z_max = max(z_hit_all.max(), z_true_all.max())
bins = np.arange(0, z_max + BIN_WIDTH, BIN_WIDTH)
centers = 0.5 * (bins[:-1] + bins[1:])

def hist(z):
    return np.histogram(z, bins=bins)[0]

counts_hit_all  = hist(z_hit_all)
counts_hit_45   = hist(z_hit_45)
counts_true_all = hist(z_true_all)
counts_true_45  = hist(z_true_45)

# ===============================
# PLOT (STEP HISTOGRAM)
# ===============================
plt.figure(figsize=(10,6))

plt.step(
    centers, counts_true_all, where="mid",
    color="black", linewidth=2.5,
    label="TRUE Z – all energies"
)

plt.step(
    centers, counts_true_45, where="mid",
    color="red", linewidth=2.5,
    label="TRUE Z – 4–5 MeV"
)

plt.step(
    centers, counts_hit_all, where="mid",
    color="black", linestyle="--", linewidth=2,
    label="HIT Z – all energies"
)

plt.step(
    centers, counts_hit_45, where="mid",
    color="red", linestyle="--", linewidth=2,
    label="HIT Z – 4–5 MeV"
)

plt.xlabel("Depth from entrance [mm]")
plt.ylabel("Number of photons per 10 mm")
plt.title("Prompt Gamma Depth Profiles\nTRUE vs HIT (Raw Counts)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# ===============================
# FILE
# ===============================
CSV_FILE = "/home/meenakshi/bragg-peak/build/detector_hits_10e8_2.csv"

# ===============================
# ENERGY WINDOW
# ===============================
E_MIN = 4.0
E_MAX = 5.0

# ===============================
# BINNING
# ===============================
BIN_WIDTH = 5.0  # mm
Z_MIN = -350
Z_MAX = 350

# ===============================
# READ CSV
# ===============================
df = pd.read_csv(CSV_FILE)

# ===============================
# FILTER 4–5 MeV GAMMAS
# ===============================
df_45 = df[
    (df["Initial_Energy_MeV"] >= E_MIN) &
    (df["Initial_Energy_MeV"] <= E_MAX)
]

print(f"4–5 MeV gammas: {len(df_45):,}")
print(f"All gammas    : {len(df):,}")

# ===============================
# EXTRACT TRUE & HIT Z (4–5 MeV)
# ===============================
z_true_45 = df_45["True_Origin_Z_mm"].values
z_hit_45  = df_45["Pos_Z_mm"].values

# ===============================
# EXTRACT TRUE & HIT Z (ALL energies)
# ===============================
z_true_all = df["True_Origin_Z_mm"].values
z_hit_all  = df["Pos_Z_mm"].values

# ===============================
# ALIGN TO COMMON ENTRANCE (TRUE, all energies)
# ===============================
z0 = z_true_all.min()

z_true_45  -= z0
z_hit_45   -= z0
z_true_all -= z0
z_hit_all  -= z0

# ===============================
# COMMON BINS
# ===============================
z_max = max(
    z_true_45.max(),
    z_hit_45.max(),
    z_true_all.max(),
    z_hit_all.max()
)

bins = np.arange(0, z_max + BIN_WIDTH, BIN_WIDTH)

# ===============================
# HISTOGRAMS
# ===============================
counts_true_45, edges = np.histogram(z_true_45, bins=bins)
counts_hit_45,  _     = np.histogram(z_hit_45,  bins=edges)

counts_true_all, _ = np.histogram(z_true_all, bins=edges)
counts_hit_all,  _ = np.histogram(z_hit_all,  bins=edges)

centers = 0.5 * (edges[:-1] + edges[1:])

# ===============================
# PLOT
# ===============================
plt.figure(figsize=(9,5))

# --- 4–5 MeV ---
plt.step(
    centers, counts_true_45,
    where="mid",
    linewidth=2.5,
    color="black",
    label="TRUE Z (4–5 MeV)"
)

plt.step(
    centers, counts_hit_45,
    where="mid",
    linewidth=2.5,
    color="red",
    label="HIT Z (4–5 MeV)"
)

# --- ALL energies ---
plt.step(
    centers, counts_true_all,
    where="mid",
    linewidth=1.8,
    linestyle="--",
    color="black",
    label="TRUE Z (all energies)"
)

plt.step(
    centers, counts_hit_all,
    where="mid",
    linewidth=1.8,
    linestyle="--",
    color="red",
    label="HIT Z (all energies)"
)

plt.xlabel("Depth from entrance [mm]")
plt.ylabel("Number of photons")
plt.title("Prompt Gamma Depth Profile\nTRUE vs HIT (4–5 MeV and All Energies)")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

E_MIN = 4.0
E_MAX = 5.0

df_45 = df[
    (df["Initial_Energy_MeV"] >= E_MIN) &
    (df["Initial_Energy_MeV"] <= E_MAX)
]

print(f"4–5 MeV photons: {len(df_45):,}")

# ---------------------------------
# HISTOGRAM (do NOT plot yet)
# ---------------------------------
counts, edges = np.histogram(
    df_45["True_Origin_Z_mm"],
    bins=1000
)

# Bin centers
centers = 0.5 * (edges[:-1] + edges[1:])

# ---------------------------------
# FIND MAXIMUM
# ---------------------------------
imax = np.argmax(counts)
z_max = centers[imax]
count_max = counts[imax]

print(f"Maximum photon count = {count_max}")
print(f"Z position of maximum = {z_max:.2f} mm")

# ---------------------------------
# PLOT
# ---------------------------------
plt.figure(figsize=(8,5))
plt.hist(
    df_45["True_Origin_Z_mm"],
    bins=1000,
    histtype="step",
    color="black"
)

plt.axvline(
    z_max,
    color="red",
    linestyle="--",
    linewidth=2,
    label=f"Max @ {z_max:.2f} mm"
)

plt.xlabel("True Z [mm]")
plt.ylabel("Number of photons")
plt.title("Depth profile of 4–5 MeV gammas")
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()
plt.show()
```

```python
# Starting point (entrance)
z_start = df_45["True_Origin_Z_mm"].min()

# Distance from start in mm
distance_mm = z_max - z_start

# Convert to cm
distance_cm = distance_mm / 10.0

print(f"Starting Z position : {z_start:.2f} mm")
print(f"Peak Z position     : {z_max:.2f} mm")
print(f"Distance from start : {distance_mm:.2f} mm = {distance_cm:.2f} cm")
```

```python
# ---------------------------------
# COMPUTE F80% (distal fall-off)
# ---------------------------------

# Peak value
peak_value = count_max
threshold_80 = 0.8 * peak_value

# Only look AFTER the peak (distal side)
counts_after = counts[imax:]
centers_after = centers[imax:]

# Find first bin where signal drops below 80%
idx_80 = np.where(counts_after <= threshold_80)[0]

if len(idx_80) == 0:
    raise RuntimeError("F80 not found — distribution may not fall off properly.")

f80_z = centers_after[idx_80[0]]

print(f"F80 (80% fall-off depth) = {f80_z:.2f} mm")
print(f"Distance from peak to F80 = {f80_z - z_max:.2f} mm")
```

```python
import matplotlib.pyplot as plt

# Select CylDetector
df_det = df[df["Detector"] == "CylDetector"]

z_true = df_det["True_Origin_Z_mm"]
z_hit  = df_det["Pos_Z_mm"]

# -------------------------------
# Split by True Z
# -------------------------------
mask_upstream   = z_true < -100
mask_downstream = z_true >= -100

# -------------------------------
# Plot
# -------------------------------
plt.figure(figsize=(6,6))

plt.scatter(
    z_true[mask_upstream],
    z_hit[mask_upstream],
    s=1,
    alpha=0.3,
    label="Upstream production",
)

plt.scatter(
    z_true[mask_downstream],
    z_hit[mask_downstream],
    s=1,
    alpha=0.3,
    label="Downstream production",
)

# Ideal line
min_z = min(z_true.min(), z_hit.min())
max_z = max(z_true.max(), z_hit.max())
plt.plot([min_z, max_z], [min_z, max_z], 'r--', label="Ideal (y = x)")

plt.xlabel("True Z (mm)")
plt.ylabel("Hit Z (mm)")
plt.title("True vs Hit Z (CylDetector)")
plt.legend(markerscale=4)
plt.tight_layout()
plt.show()
```
